```python
import seaborn as sns
sns.set_theme(style="whitegrid",font_scale=1.3,rc={"grid.linewidth": 0.5,'lines.linewidth': 5,})

import sympy
import numpy as np
```

```python
from sympy import symbols
from sympy.plotting import plot

x = symbols('x')
y = symbols('y')
```

## Example 1.0

```python
f = x + 2
g = x - 2

x_array = np.linspace(0, 3)
f_array = sympy.lambdify(x, f)(x_array)
g_array = sympy.lambdify(x, g)(x_array)

fb = {'x':x_array,'y1':f_array,'y2':g_array,'facecolor' : 'green', 'alpha' : 0.2}

i0=plot(f,g,
     ylim=(-5,8),
     xlim=(-3.5,3.5),
     size=(8,6),
     ylabel='',
     fill = fb
    )

i0.save('images/curve0.png')
```
